```python
import pandas as pd

df = pd.read_csv("custmore_data/telco-customer-churn-on-icp4d-master/data/Telco-Customer-Churn.csv")

print(df.head())
```

```python
print(df.shape)
```

```python
df.info()    #Data types  , Missing values
```

***Check its type — you’ll notice it’s object (string), which is wrong.***

This converts it to numeric
👉 Invalid values become NaN

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
df = df.dropna()   # missing values 
```

```python
df.info()
```

```python
print(df["TotalCharges"].dtype)
```

```python
df = df.drop("customerID", axis=1)
```

```python
df["Churn"] = df["Churn"].map({"Yes": 1, "No": 0})    # convert churn
```

***Encoding***

convert categorical data into numerical 

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.info()
```

```python
print(df.head())
print(df.shape)
```

***Define X and y***

```python
x = df.drop("Churn", axis =1)    # input features 
y = df["Churn"]     # y = target(what we predict)
```

***Train-Test Split***

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(
    x, y, test_size=0.2, random_state= 42
    
)
```

***Train Model***

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

***Prediction***

```python
y_pred = model.predict(x_test)
```

***Evaluation***

```python
from  sklearn.metrics import accuracy_score, confusion_matrix, classification_report


print("Accuracy:",accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

***Random Forest***

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train, y_train)

y_pred_rf = rf.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

***classification_report:***

Instead of just accuracy, it shows:

Precision
Recall
F1-score
Support (number of samples)

“I started with Logistic Regression as a baseline model. Then I applied Random Forest to capture non-linear patterns and improve performance. I compared both models using classification metrics.”
